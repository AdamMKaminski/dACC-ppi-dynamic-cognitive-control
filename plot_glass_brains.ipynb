```python
# plot seeds
# Fronto-parietal affiliated seed
nii_file = f'{path}/conflict_cont.nii'
img = nib.load(nii_file)
data = img.get_fdata()
binary_data = np.where(data != 0, 1, 0)
binary_data = binary_data.astype(np.uint8)
binary_img = nib.Nifti1Image(binary_data, img.affine)

colors = ['black', 'orange']  
cmap = ListedColormap(colors)

plotting.plot_glass_brain(binary_img, plot_abs=True, title='', colorbar=False, cmap=cmap)
plotting.show()
```

```python
# Salience affiliated seed
nii_file2 = f'{path}/conflict_salventattn.nii'
img2 = nib.load(nii_file2)
data2 = img2.get_fdata()
binary_data2 = np.where(data2 != 0, 1, 0)
binary_data2 = binary_data2.astype(np.uint8)
binary_img2 = nib.Nifti1Image(binary_data2, img2.affine)

colors = ['black', 'violet']  
cmap = ListedColormap(colors)

plotting.plot_glass_brain(binary_img2, plot_abs=True, title='', colorbar=False, cmap=cmap)
plotting.show()
```
